```python
import pandas as pd
loan = pd.read_csv("loan_data_set.csv")
```

```python
loan
```

Dropping variables we do not need or won't use for this project.

```python
loan = loan.drop(['Loan_ID','Dependents', 'Education','CoapplicantIncome','Loan_Amount_Term'], axis=1)
loan
```

Removing missing values

```python
loan = loan.dropna()
```

Removing Outliers

```python
Q1 = loan.quantile(0.25)
Q3 = loan.quantile(0.75)
IQR = Q3 - Q1

loan = loan[~((loan < (Q1 - 1.5 * IQR)) |(loan > (Q3 + 1.5 * IQR))).any(axis=1)]
```

Encoding to switch categorical data to numerical

```python
loan = pd.get_dummies(loan, columns = ['Gender', 'Married', 'Self_Employed', 'Property_Area','Loan_Status'])
loan = loan.drop(['Loan_Status_N', 'Married_No','Self_Employed_No', 'Gender_Male', 'Property_Area_Semiurban'], axis = 1)
loan
```

Renames columes

```python
loan.columns = ['ApplicantIncome', 'LoanAmount', 'Credit_History', 'Gender', 'Married','Self_Employed', 'Propert_Area_Rural', 'Property_Area_Urban', 'Loan_Status']
loan
```

Define X and Y features

```python
X = loan.drop(["Loan_Status"], axis=1)
y = loan["Loan_Status"]
```

Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

**Models**

Baseline Classification

```python
from sklearn.dummy import DummyClassifier

# create dummy classifier
loan_dummy_clf = DummyClassifier(strategy='uniform', random_state=42)

# train and fit a model
loan_dummy_clf.fit(X_train, y_train)

# get accuracy score
loan_dummy_clf.score(X_test, y_test)
```

Linear Perceptron

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create a Linear Perceptron model
loan_perceptron_model = Perceptron(tol=1e-3, random_state=0)

#train and fit the model
loan_perceptron_model.fit(X, y)

#set your model to predict x
loan_predicted_labels = loan_perceptron_model.predict(X)

#get accuracy score
loan_training_accuracy = accuracy_score(y, loan_predicted_labels)
print(f"Accuracy Score: {loan_training_accuracy}")
```

LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

#create a logistic regression model
loan_model = LogisticRegression()

#fit and train the model
loan_model.fit(X,y)

#get accuracy score
loan_model.score(X,y)
```

SVM

```python
from sklearn.svm import SVC

#train the model in a linear kernel
loan_model = SVC(kernel = "linear")

#fit the model
loan_model.fit(X,y)

#get accuracy score
loan_model.score(X,y)
```

DecisionTreeClassifier

```python
from sklearn import tree
from sklearn import metrics

#create and train DecisionTreeClassifier
loan_tree_clf= tree.DecisionTreeClassifier()

loan_tree_clf.fit(X_train, y_train)

# make prediction
loan_tree_pred_Y = loan_tree_clf.predict(X_test)

#get accuracy score
print("Accuracy: {}".format(metrics.accuracy_score(y_test, loan_tree_pred_Y)))
```

RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Create and Train a RandomForestClassifier
loan_rf_clf = RandomForestClassifier()
loan_rf_clf.fit(X_train, y_train)

# Make predictions
loan_rf_predictions = loan_rf_clf.predict(X_test)

# get accuracy score
print("Accuracy: {}".format(metrics.accuracy_score(y_test, loan_rf_predictions)))
```

K-NN

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#create array to hold accuracy score
KNN_accuracy_score = []

#for loop to train and fit KNeighborsClassifier between k=1-25
for k in range(1, 25):
    loan_knn = KNeighborsClassifier(n_neighbors=k)
    loan_knn.fit(X_train, y_train)
    y_pred = loan_knn.predict(X_test)
    loan_knn_accuracy = accuracy_score(y_test, y_pred)
    KNN_accuracy_score.append(loan_knn_accuracy)
    print(f"K = {k}, Accuracy: {loan_knn_accuracy}")


#print out best k and accruacy scores
loan_best_k = range(1,25)[np.argmax(KNN_accuracy_score)]
loan_best_accuracy = max(KNN_accuracy_score)
print(f"The best k is {loan_best_k} with an accuracy of {loan_best_accuracy:.2f}")
```

**Hyperparameter Tuning from Preliminary Results**

Increasing Logistic Regression Decision Threshold

```python
pred_x = loan_model.predict_proba(X)
threshold = 0.6
result = (pred_x[:,1] > threshold).astype(int)
accuracy = sum(result == y)/len(y)
print(f"Accuracy with decision threshold 0.6 is",{accuracy})

pred_x = loan_model.predict_proba(X)
threshold = 0.7
result = (pred_x[:,1] > threshold).astype(int)
accuracy = sum(result == y)/len(y)
print(f"Accuracy with decision threshold 0.7 is",{accuracy})

pred_x = loan_model.predict_proba(X)
threshold = 0.8
result = (pred_x[:,1] > threshold).astype(int)
accuracy = sum(result == y)/len(y)
print(f"Accuracy with decision threshold 0.8 is",{accuracy})

pred_x = loan_model.predict_proba(X)
threshold = 0.9
result = (pred_x[:,1] > threshold).astype(int)
accuracy = sum(result == y)/len(y)
print(f"Accuracy with decision threshold 0.9 is",{accuracy})
```

Removing Initital Support Vectors

```python
X = np.delete(X, loan_model.support_, 0)
y = np.delete(y, loan_model.support_, 0)

loan_model = SVC(kernel ="linear")
loan_model.fit(X, y)
loan_model.score(X,y)
```

Testing Additional Values of K

```python
loan_best_k = range(25,100)[np.argmax(KNN_accuracy_score)]
loan_best_accuracy = max(KNN_accuracy_score)
print(f"The best k is {loan_best_k} with an accuracy of {loan_best_accuracy:.2f}")
```

Maximum Depth in Decision Tree

```python
#create and train DecisionTreeClassifier
loan_tree_clf= tree.DecisionTreeClassifier(max_depth=3)

loan_tree_clf.fit(X_train, y_train)

# make prediction
loan_tree_pred_Y = loan_tree_clf.predict(X_test)

#get accuracy score
print("Accuracy: {}".format(metrics.accuracy_score(y_test, loan_tree_pred_Y)))
```

Number of Estimators for Random Forest

```python
# Create and Train a RandomForestClassifier
loan_rf_clf = RandomForestClassifier(n_estimators=10)
loan_rf_clf.fit(X_train, y_train)

# Make predictions
loan_rf_predictions = loan_rf_clf.predict(X_test)

# get accuracy score
print("Accuracy: {}".format(metrics.accuracy_score(y_test, loan_rf_predictions)))

# Create and Train a RandomForestClassifier
loan_rf_clf = RandomForestClassifier(n_estimators=250)
loan_rf_clf.fit(X_train, y_train)

# Make predictions
loan_rf_predictions = loan_rf_clf.predict(X_test)

# get accuracy score
print("Accuracy: {}".format(metrics.accuracy_score(y_test, loan_rf_predictions)))
```
